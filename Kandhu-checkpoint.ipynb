# Missing values

```python
missing_values = pd.DataFrame({'No of missing values': df.isnull().sum(), 
                              '%  of missing values': (df.isnull().sum()/len(df))*100})
missing_values
```

```python
# Handle missing data points with Linear Interpolation

for column in df.columns:
    if df[column].isnull().any():
        df[column] = df[column].interpolate(method='linear')
print(df.isnull().sum())
```

```python
# Handle missing data points for first rows
# Filling NaN values in the first row with the mean of the 2nd and 3rd rows

if df.iloc[0].isnull().any():
    mean_val = df.iloc[1:3].mean()
    df.iloc[0] = df.iloc[0].fillna(mean_val)
    
print(df.isnull().sum())
```

# Split dataset into train and test

```python
x = df1.drop(['SON'],axis=1)
y = df1['SON']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size = 0.7,random_state =42)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

# Linear regression

```python
lr =  LinearRegression().fit(xtrain,ytrain)

Y_pred_train_lr = lr.predict(xtrain)

Y_pred_test_lr = lr.predict(xtest)
```

```python
# r2_score for lr
```

```python
r2_test_lr = r2_score(ytest,Y_pred_test_lr)
r2_test_lr
```

```python
r2_train_lr = r2_score(ytrain,Y_pred_train_lr)
r2_train_lr
```

```python
# rmse for lr
```

```python
mse_test_lr = mean_squared_error(ytest,Y_pred_test_lr)
mse_test_lr
```

```python
rmse_test_lr = np.sqrt(mse_test_lr)
rmse_test_lr
```

# Ridge

```python
rg =  Ridge().fit(xtrain,ytrain)

Y_pred_train_rg = rg.predict(xtrain)

Y_pred_test_rg = rg.predict(xtest)
```

```python
# r2_score for rg
```

```python
r2_test_rg = r2_score(ytest,Y_pred_test_rg)
r2_test_rg
```

```python
r2_train_rg = r2_score(ytrain,Y_pred_train_rg)
r2_train_rg
```

```python
# rmse for rg
```

```python
mse_test_rg = mean_squared_error(ytest,Y_pred_test_rg)
mse_test_rg
```

```python
rmse_test_rg = np.sqrt(mse_test_rg)
rmse_test_rg
```

# lasso
